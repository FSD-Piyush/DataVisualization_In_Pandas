# 1.Series

```python
import pandas as pd

x = [3,4,5,6,7,8]

var = pd.Series(x)

print(var)
print(type(var))

print(var[2])
```

```python
x = [3,4,5,6]

var = pd.Series(x,index=['a','s','d','f'])

print(var)
print(type(var))

print(var[2])
```

```python
x = [3,4,5,6]

var = pd.Series(x,index=['a','s','d','f'],dtype="float")

print(var)
print(type(var))

print(var[2])
```

```python
x = [3,4,5,6]

var = pd.Series(x,index=['a','s','d','f'],dtype="float",name="python")

print(var)
print(type(var))

print(var[2])
```

```python
dic = {"name":['python','c','c++','java'],"por":[12,13,14,15],"rank":[1,4,3,2]}

var1 = pd.Series(dic)

# Either print(var1) or simply var1, both will work.
var1
```

```python
s = pd.Series(12,index=[1,2,3,4,5,6,7])
print(s)
print(type(s))
```

```python
s1 = pd.Series(12,index=[1,2,3,4,5,6,7])
s2 = pd.Series(12,index=[1,2,3,4])

print(s1+s2)
# This broadcasting error/method not possible in numpys. So, we do here in pandas.
```

# 10.Merging & Concat

```python
var3 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var4 = pd.DataFrame({"A":[1,2,3,4],"C":[21,22,23,24]})

pd.merge(var3,var4,on="A")
```

```python
var3 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var4 = pd.DataFrame({"A":[1,2,3,4],"C":[21,22,23,24]})

pd.merge(var4,var3,on="A")
```

```python
var3 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var4 = pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})

pd.merge(var3,var4,on="A")
```

```python
var3 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var4 = pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})

pd.merge(var3,var4, how = "right")
```

```python
var3 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var4 = pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})

pd.merge(var3,var4, how = "left")
```

```python
var3 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var4 = pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})

pd.merge(var3,var4, how = "outer")
```

```python
var3 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var4 = pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})

pd.merge(var3,var4, how = "outer", indicator=True)
```

```python
var3 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var4 = pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})

pd.merge(var3,var4,left_index=True,right_index=True)
```

```python
var3 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var4 = pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})

pd.merge(var3,var4,left_index=True,right_index=True,suffixes=("_name","_id"))
```

```python
sr1 = pd.Series([1,2,3,4])
sr2 = pd.Series([11,21,31,41])

pd.concat([sr1,sr2])
```

```python
d1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 = pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})

pd.concat([d1,d2])
```

```python
d1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 = pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})

pd.concat([d1,d2],axis=1)
```

```python
d1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 = pd.DataFrame({"A":[1,2],"B":[21,22]})

pd.concat([d1,d2],axis=1,join="outer")
```

```python
d1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 = pd.DataFrame({"A":[1,2],"B":[21,22]})

pd.concat([d1,d2],axis=1,join="inner")
```

# 12.Join & Append() 

```python
var6 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var7 = pd.DataFrame({"C":[10,20,30,40],"D":[11,22,33,44]})

var6.join(var7)
```

```python
var6 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var7 = pd.DataFrame({"C":[10,20],"D":[11,22]})

var6.join(var7)
```

```python
var6 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
var7 = pd.DataFrame({"C":[10,20],"D":[11,22]})

var6.join(var7)
```

```python
var6 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
var7 = pd.DataFrame({"C":[10,20],"D":[11,22]},index=["a","b"])

var6.join(var7)
```

```python
var6 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
var7 = pd.DataFrame({"C":[10,20],"D":[11,22]},index=["a","b"])

var7.join(var6)
```

```python
var6 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
var7 = pd.DataFrame({"C":[10,20],"D":[11,22]},index=["a","b"])

var7.join(var6,how="left")
```

```python
var6 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
var7 = pd.DataFrame({"C":[10,20],"D":[11,22]},index=["a","b"])

var7.join(var6,how="right")
```

```python
var6 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
var7 = pd.DataFrame({"C":[10,20],"D":[11,22]},index=["a","b"])

var7.join(var6,how="outer")
```

```python
var6 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
var7 = pd.DataFrame({"C":[10,20],"D":[11,22]},index=["a","b"])

var7.join(var6,how="inner")
```
